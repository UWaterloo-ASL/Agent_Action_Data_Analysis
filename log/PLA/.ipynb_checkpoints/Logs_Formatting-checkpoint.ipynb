## This file is to formulate the logs. Particularly, it is for 2018-09-21 and 2018-09-24.

```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
logs = glob.glob('*.csv')
logs.sort()

interaction_data = {}
for file in logs:
    print(file)
    interaction_data[file] = pd.read_csv(file, header=None)
```

### delete blank rows in 2018-09-21.csv

```python
for index, row in interaction_data['2018-09-21.csv'].iterrows():
    if all(row.isnull()):
        interaction_data['2018-09-21.csv'] = interaction_data['2018-09-21.csv'].drop(index)
        
```
